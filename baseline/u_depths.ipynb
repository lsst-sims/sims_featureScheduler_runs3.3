```python
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
import pandas as pd
import sqlite3
```

```python
sql = 'select fivesigmadepth,airmass,skybrightness from observations where filter="u";'

con = sqlite3.connect('baseline_v3.3_10yrs.db')
df33 = pd.read_sql(sql, con)
con.close()



con = sqlite3.connect('baseline_v3.2_10yrs.db')
df32 = pd.read_sql(sql, con)
con.close()

```

```python
_temp = plt.hist(df33['fiveSigmaDepth'])
_temp = plt.hist(df32['fiveSigmaDepth'])
```

```python
np.median(df33['fiveSigmaDepth']) - np.median(df32['fiveSigmaDepth'])
```

```python
_temp = plt.hist(df33['skyBrightness'])
_temp = plt.hist(df32['skyBrightness'])
```

```python
np.median(df33['skyBrightness']) - np.median(df32['skyBrightness'])
```
